### Exercise 4: Making some plots

Now that you are used to the basic functionality, we can get started with some visualizations.

**4.a: Make a plot of Germany's life expectancy vs. population over time**.

That is, select the data for Germany, and plot all values of life expectancy against population size, connected by lines.

````python
de_data = data[data['country'] == 'Germany']

plt.figure(figsize=(5, 5))
plt.plot(de_data['pop'], de_data['lifeExp'], c='k')
plt.xlabel('Population')
plt.ylabel('Life Expectancy')
plt.title('German Life Expectancy vs Population')
plt.show()
````

You will see a distinctive jump upwards in your plot. Can you guess where this comes from? There are actually two separate events at work, here. To get a better idea, try plotting life expectancy and population separately against time. 

````python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(de_data['year'], de_data['lifeExp'], c='k')
plt.ylabel('Life Expectancy')
plt.xlabel('Year')

plt.subplot(1, 2, 2)
plt.plot(de_data['year'], de_data['pop'], c='k')
plt.ylabel('Population')
plt.xlabel('Year')

plt.suptitle('German Life Expectancy & Population over time')
plt.show()
````

You will see that life expectancy grew relatively linearly, while the population development has a *kink* around 1965-1970 that flattened it out, and another kink in 1990. These two events [had](https://de.wikipedia.org/wiki/Pillenknick) separate [causes](https://en.wikipedia.org/wiki/German_reunification).

### Exercise 5: Making useful plots

This final exercise is a bit longer. It's goal is to give a first intuition for what makes a good plot. There will be more to come on this topic later in the course.

**5.a: Make a plot that, for all countries, plots gdp per capita agains the year**

````python
countries = list(set(data['country']))

plt.figure(figsize=(10, 5))

for country in countries:
  country_data = data[data['country'] == country]
  plt.plot(country_data['year'], country_data['gdpPercap'], label=country)

plt.title(f"GDP per capita of {len(countries)} countries against time")
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
````

as you will see, this is not a particularly helpful plot. Let's see if we can improve it a bit.

**5.b: Make 5 separate plots in one row, one for each continent, with one line per country, plotting gdp per capita against the year.**  Add labels to the axis and title each plot with the name of the continent. Make sure each plot uses the same limits on abscissa and ordinal.

````python
continents = list(set(data['continent']))

fig, axs = plt.subplots(
    nrows=1,
    ncols=len(continents),
    sharex=True,
    sharey=True,
    figsize=(len(continents)*5, 5)
)


for continent_idx, continent in enumerate(continents):
  continent_data = data[data['continent'] == continent]
  continent_countries = list(set(continent_data['country']))

  ax = axs[continent_idx]

  for country in countries:
    country_data = continent_data[continent_data['country'] == country]

    ax.plot(
        country_data['year'],
        country_data['gdpPercap'],
        label=country
    )

    ax.set_xlabel('Year')
    ax.set_ylabel('Population')
    ax.set_title(continent)

plt.suptitle(f"GDP per capita of {len(continents)} continents against time")
plt.show()
````

**5.c: To improve the plot further, plot the GDP values on a log scale. Also compute the mean of each continent and plot it alongside the countries in each panel.** Your final plot might look like this:

![title](../data/gdp_by_continent.svg)

````python
continents = list(set(data['continent']))

fig, axs = plt.subplots(
    nrows=1,
    ncols=len(continents),
    sharex=True,
    sharey=True,
    figsize=(len(continents)*5, 5)
)


for continent_idx, continent in enumerate(continents):
  continent_data = data[data['continent'] == continent]
  continent_countries = list(set(continent_data['country']))

  ax = axs[continent_idx]

  for country in countries:
    country_data = continent_data[continent_data['country'] == country]

    ax.plot(
        country_data['year'],
        country_data['gdpPercap'],
        label=country,
        c='#666666'
    )

  mean_continent_data = continent_data.groupby('year').mean()
  ax.plot(
        mean_continent_data.index,
        mean_continent_data['gdpPercap'],
        c='#ee3333',
        lw=3
    )

  ax.set_xlabel('Year')
  ax.set_ylabel('Population')
  ax.set_title(continent)
  ax.grid(True)
  ax.set_yscale('log')

plt.suptitle(f"GDP per capita of {len(continents)} continents against time")
plt.show()
````
